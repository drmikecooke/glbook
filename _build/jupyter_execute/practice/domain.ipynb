````python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
from glcd import plt1D,gl1D,nat,SC,nrm

h0=1/2**(1/2)

for k in range(-1,2):
  kappa=2**k
  plt1D(kappa,nrm(0,40,h0),SC(10/kappa**(1/2)),f"\kappa={kappa}")

plt.legend()
plt.show()
````

````python
def dG1ABA(kappa,h0,A,B):
  sol=gl1D(kappa,A,B)
  def g(x):
    f,df,a,da=sol.sol(x)
    return -f**4/2+(da-h0)**2  
  return quad(g,A[0],B[0])

h0=1/2**(1/2)
Kappa=np.linspace(.6,1,40)
plot=[dG1ABA(kappa,h0,nrm(-4,40,h0),SC(20))[0] for kappa in Kappa]
plt.plot(Kappa,plot)
plt.xlabel("$\kappa$")
plt.ylabel("$\delta$")
plt.title(f"$h_0={h0:.2f}$")
plt.show()
````

The $\delta$ scaled energy quantity going negative for high $\kappa$ indicates that &lsquo;domain walls&rsquo; become energeticallly favourable, allowing regions of flux to penetrate the superconductor. This is known as &lsquo;Type II&rsquo; behavior. For low $\kappa$ one has a &lsquo;Type I&rsquo; superconductor, where the no flux penetration is allowed, and the critical field $h_c$ is sharp at $1/\sqrt2$ for the scaled theory.

## $\kappa=1/\sqrt2$

When $\kappa=1/\sqrt2$, if we assume the difference between the superconducting and normal states is zero:

$$-\frac{\psi^4}{2}+(h_c-b)^2=0$$

we have:

$$\psi^2=\sqrt2(h_c-b)=1-\sqrt2b$$(b)

Differentiating:

$$2\psi\psi'=-\sqrt2a''=-\sqrt2\psi^2 a\implies \psi'=-\frac{\psi a}{\sqrt2}$$(psip)

using the GL equation for $a''$.

Again:

$$\psi''=-\frac{\psi'a+a'\psi}{\sqrt2}=\frac12(a^2+\psi^2-1)\psi$$

This is the GL equation for $\kappa=1/\sqrt2$. The final step uses the equations  {eq}`psip` and {eq}`b` to replace $\psi'$ and $b=a'$, respectively.
